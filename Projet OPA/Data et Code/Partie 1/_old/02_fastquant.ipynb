```python
import requests
import pandas as pd
```

```python
cle_api = "DUA7M7J7PB3YFCYK"
```

```python
def get_crypto_price(cle_api,symbol, exchange, start_date = None):
    api_url = f'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol={symbol}&market={exchange}&apikey={cle_api}'
    print(api_url)
    raw_df = requests.get(api_url).json()
    df = pd.DataFrame(raw_df['Time Series (Digital Currency Daily)']).T
    df = df.rename(columns = {'1a. open (USD)': 'open', '2a. high (USD)': 'high', '3a. low (USD)': 'low', '4a. close (USD)': 'close', '5. volume': 'volume'})
    for i in df.columns:
        df[i] = df[i].astype(float)
    df.index = pd.to_datetime(df.index)
    df = df.iloc[::-1].drop(['1b. open (USD)', '2b. high (USD)', '3b. low (USD)', '4b. close (USD)', '6. market cap (USD)'], axis = 1)
    if start_date:
        df = df[df.index >= start_date]
    return df
```

```python
btc = get_crypto_price(cle_api, symbol = 'BTC', exchange = 'USD', start_date = '2023-01-01')
btc
```
